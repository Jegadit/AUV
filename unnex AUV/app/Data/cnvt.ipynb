```python
import pandas as pd
```

```python
data = pd.read_csv('SeaDragonData_2981.csv')
data.head()
```

```python
newdf = pd.DataFrame()
```

```python
lines = data['payload']

# Parse each line into a dictionary
data_dicts = [eval(line) for line in lines]

# Accessing each element
for data_dict in data_dicts:
    internal_temp = data_dict['internal_temperature']['N']
    pressure = data_dict['pressure']['N']
    external_temp = data_dict['external_temperature']['N']
    timestamp = data_dict['timestamp']['N']

    print("Internal Temperature:", internal_temp)
    print("Pressure:", pressure)
    print("External Temperature:", external_temp)
    print("Timestamp:", timestamp)
    print()
```

```python
for line in lines:
    data_dict = eval(line)
    row = {
        'internal_temperature': float(data_dict['internal_temperature']['N']),
        'pressure': float(data_dict['pressure']['N']),
        'external_temperature': float(data_dict['external_temperature']['N']),
        'timestamp': int(data_dict['timestamp']['N'])
    }
    newdf = newdf.append(row, ignore_index=True)
```
